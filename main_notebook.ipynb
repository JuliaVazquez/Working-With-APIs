# Spanish Holidays 2021

```python
import getpass
apikey = getpass.getpass()
```

```python
holydays = requests.get("")
print("Spanish Holydays:", holydays.status_code)
```

```python
response = requests.get('https://holidayapi.com/v1/holidays?pretty&country=ES&year=2021&key='+apikey)
response
```

```python
results = response.json()
# results
```

```python
# results['holidays']
```

```python
holidays = pd.DataFrame(results["holidays"]).reset_index(drop=True)
holidays
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
flat1 = flatten(holidays,['weekday'])
flat1.head()
```

```python
flat2 = flatten(flat1,['weekday_date','weekday_observed'])
flat2.head()
```

```python
holidays_df = flat2.drop(['uuid'],axis=1)
holidays_df
```
